````python
class LPAmPlus:
    def __init__(self, graph: nx.Graph):
        """
        # Arguments:
        - graph: nx.Graph, the graph on which the algorithm will be applied
        """
        self.graph: nx.Graph = graph

        """
        labels gives all the communities present in the network
        Assign an initial label to each node
        """
        self.labels: Iterable[int] = np.arange(len(graph.nodes))
        labels = self.labels
        for node in self.graph.nodes :
            self.graph.nodes[node]["label"] = labels[node]
        

        """
        Optional: Save your labels to visualize the evolution of communities
        1 dict for each step of the algorithm in the format 
        [
            (title_step1, {node_idx1: label, ...}),
            (title_step2, {..., ...}),
        ]
        """
        self.labels_history: List[Tuple[str, Dict[int, int]]] = []

    def modularity(self) -> float:
        """
        Function that calculates the current modularity of the network
        nx.linalg.modularity_matrix can be used

        # Arguments:
        - None (the network evaluated is self.graph)

        # Output (see explanation Q1.2)
                    n
            1     ___                                                                               degree(u) ⋅ degree(v)
        Q = ── ⋅   ╲    B   ⋅ δ(lu, lv) with B   = A   - P  , A   = 1 if edge between u and v  P   = ─────────────────────
            2m     ╱     uv                  uv    uv    uv   uv                               uv           2m
                  ‾‾‾
                u,v = 1

        - float, the modularity of the network
        """
        G = self.graph
        m = len(G.edges)
        B = nx.linalg.modularity_matrix(G)
        Q = 0
        for u in G.nodes :
            for v in G.nodes :
                if G.nodes[u]['label'] == G.nodes[v]['label'] :
                    Q += B[u][v]
        return Q / (2 * m)

    def label_evaluation(self, current_node: int, new_label: int) -> float:
        """
        Term to optimize when replacing labels

        # Arguments
        - current_node: int, the node id to evaluate
        - new_label: int, the potential new_label

        # Output
            n
            ___
            ╲    B   ⋅ δ(lu, l) with l=new_label
            ╱     ux
            ‾‾‾
            u = 1
        - float, the modularity gain brought by this assignement
        """
        G = self.graph
        B = nx.linalg.modularity_matrix(G)
        sum = 0
        for u,node_u in enumerate(G.nodes):
            if node_u == current_node:
                continue
            if G.nodes[u]["label"] == new_label :
                sum += B[u][current_node]
        return sum


    def update_label(self, current_node: int) -> None:
        """
        This function chooses the new label for the current node specified.
        We are in the case where we choose to update the label of `current_node` and as the paper says we are
        "selecting a new label that maximizes the second term on the right-hand side of (5)"
        "If more than one label are the most frequent ones, the new label is chosen randomly from them."

        /!\ Don't forget to remove the unused labels from the `labels` attribute.
        /!\ The best label can be the node's current label.

        # Arguments
        - current_node: int, the node id to evaluate
        """
        G = self.graph
        labels = self.labels
        current_label = G.nodes[current_node]["label"]

        #Find best label
        best_score = self.label_evaluation(current_node, current_label)
        best_labels = [current_label]
        for label in labels :
            score = self.label_evaluation(current_node, label)
            if score > best_score :
                best_score = score
                best_labels = [label]
            elif score == best_score :
                best_labels.append(label)

        new_label = random.choice(best_labels)

        if new_label != current_label :
            
            G.nodes[current_node]["label"] = new_label

            #Delete unused labels
            used = False
            for u in G.nodes :
                if G.nodes[u]["label"] == current_label :
                    used = True
            if not used :
                new_labels = np.delete(labels, np.where(labels == current_label))
                self.labels = new_labels
        

    def LPAm(self) -> None:
        """
        Function that applies the LPAm algorithm on the network
        Same principle as LPA
        "each node sequentially updates its label to a new one" (article)
        But using LPAm
        "Barber and Clark extend LPA by modifying the label updating rule so that modularity can be maximized, and propose a new algorithm called LPAm [6]..."
        """

        G = self.graph
        Q = self.modularity()

        n = G.number_of_nodes()
        i = 0
        step = 1

        while i < n :

            for node in tqdm(G.nodes) : 
                self.update_label(node)
                new_Q = self.modularity()
                if (new_Q == Q) :
                    i +=1
                elif (new_Q > Q) :
                    i =0
                else :
                    print("Erreur : la modularité a diminué")
                    return
                Q = new_Q

            step += 1


    def merge_communities(self) -> bool:
        """
        Function that find the combination of communities that maximize the modularity gain
        and combines them together

        # Out
        - bool, False if no combination increase the modularity, True otherwise
        """
        init_Q = self.modularity()
        label_duos = list(combinations(self.labels, 2))
        comb_Q_dict = {}
        for duo in label_duos :
            original_graph = copy.deepcopy(self.graph)
            original_labels = copy.deepcopy(self.labels)
            
            for node in self.graph.nodes :
                if self.graph.nodes[node]["label"] == duo[0] :
                    self.graph.nodes[node]["label"] = duo[1]

            duo_Q = self.modularity()
            if duo_Q >= init_Q :
                comb_Q_dict[duo] = duo_Q
            self.graph = original_graph
            self.labels = original_labels

        if comb_Q_dict == {} :
            return False
        else : 
            best_duo = max(comb_Q_dict, key=comb_Q_dict.get)
            for node in self.graph.nodes :
                if self.graph.nodes[node]["label"] == best_duo[0] :
                    self.graph.nodes[node]["label"] = best_duo[1]
            new_labels = np.delete(self.labels, np.where(self.labels == best_duo[0]))
            self.labels = new_labels
            return True
        

    def find_communities(self) -> None:
        """
        Function that applies the LPAm+ algorithm on the network using LPAm and merge_communities
        """
        self.labels_history.append(("Etat initial", {node : self.graph.nodes[node]["label"] for node in self.graph.nodes}))
        merge = True
        i = 1
        while merge == True :
            print(i, self.modularity())
            self.LPAm()
            self.labels_history.append((f"LPAm {i}", {node : self.graph.nodes[node]["label"] for node in self.graph.nodes}))
            merge = self.merge_communities()
            self.labels_history.append((f"merge {i}", {node : self.graph.nodes[node]["label"] for node in self.graph.nodes}))
            i += 1
````

````python
def load_unweighted_network(source_csv: str) -> nx.Graph:
    """Function that loads from a csv a networkx graph
    
    # Arguments
    - source_csv: str, the path to the csv that allows to build the graph
    
    # Output
    - nx.Graph, the networkx graph built from the csv
    """
    df = pd.read_csv(source_csv, sep=";")
    nodes = df["Hero_id"].unique()
    G = nx.Graph()
    G.add_nodes_from(nodes)

    nodes_comb = list(combinations(nodes, 2))
    for comb in nodes_comb :
        movies1 = df.loc[df["Hero_id"] == comb[0], "Movie"].tolist()
        movies2 = df.loc[df["Hero_id"] == comb[1], "Movie"].tolist()
        if any(i in movies1 for i in movies2) :
            G.add_edge(comb[0], comb[1])

    return G
````

A first simple measure to find the importance of a node in a network is the degree centrality. It is calculated $$C_{D}(i) = \frac{degree(i)}{n-1}$$

- i: a node in the network
- n: the number of nodes
- degree: the number of edges attached to the node

````python
def calculate_degree_centrality(social_network: nx.Graph):
    """Calculate the degree centrality of each node in social_network and insert the value of each node in a new attribute"""
    n = len(social_network.nodes)
    for node in social_network.nodes :
        C_d = social_network.degree[node] / (n-1)
        social_network.nodes[node]['degree centrality'] = C_d

field_degree_centrality = "degree centrality"

def test_calculate_degree_centrality():
    """Tests the function calculate_degree_centrality"""
    if field_degree_centrality == "":
        raise ValueError("Fill the variable field_degree_centrality with the field where for each node we can find its centrality")
    
    social_network = load_unweighted_network(str(ROOT_DATA / "tp-marvel-dataset.csv"))
    
    calculate_degree_centrality(social_network)
    centralities = [[social_network.nodes[i][field_degree_centrality],i] for i in range(social_network.number_of_nodes())]
    exp_centralities = [[v,k] for k,v in nx.degree_centrality(social_network).items()]
    centralities.sort(key=lambda x:-x[0])
    exp_centralities.sort(key=lambda x:-x[0])
    tolerance = 0.001
    for i,((c_pred, idx_pred), (c_true, idx_true)) in enumerate(zip(centralities, exp_centralities)):
        assert idx_pred==idx_true, f"Expecting the {i}-st highest degree centrality node to be {idx_true}"
        assert np.isclose(c_pred,c_true, rtol=tolerance, atol=tolerance), f"Expecting the {i}-st highest degree centrality to be {c_true}"
    
if NOTEBOOK:
    test_calculate_degree_centrality()
````

````python
def calculate_closeness_centrality(social_network: nx.Graph):
    for node in social_network.nodes :
        dist = 0
        connected = nx.shortest_path(social_network, source=node).keys()

        for target in connected :
            if target != node :
                dist += nx.shortest_path_length(social_network, source=node, target=target)
        C_p = 1 / (dist / (len(connected) - 1))
        social_network.nodes[node]['closeness centrality'] = C_p
            
  
        
field_closeness_centrality = "closeness centrality"

def test_hint_calculate_closeness_centrality():
    if field_closeness_centrality == "":
        raise ValueError("Fill the variable field_closeness_centrality with the field where for each node we can find its centrality")
    social_network = load_unweighted_network(str(ROOT_DATA / "tp-marvel-dataset.csv"))
    calculate_closeness_centrality(social_network)
    centralities = [[social_network.nodes[i][field_closeness_centrality],i] for i in range(social_network.number_of_nodes())]
    exp_centralities = [[v,k] for k,v in nx.closeness_centrality(social_network,wf_improved=False).items()]
    centralities.sort(key=lambda x:-x[0])
    exp_centralities.sort(key=lambda x:-x[0])
    tolerance = 0.001
    for i,((c_pred, idx_pred), (c_true, idx_true)) in enumerate(zip(centralities, exp_centralities)):
        assert idx_pred==idx_true, f"Expecting the {i}-st highest closeness centrality node to be {idx_true}"
        assert np.isclose(c_pred,c_true, rtol=tolerance, atol=tolerance), f"Expecting the {i}-st highest closeness centrality to be {c_true}"
    
if NOTEBOOK:
    test_hint_calculate_closeness_centrality()
````

A final simple measure to find the importance of a node in a network is the betweeness centrality. It is calculated $$C_{I}(i) = \frac{\sum_{j<k}f_{jk}(i)}{\binom{n-1}{2}}$$

- n: the number of nodes in the network
- i: a node in the network
- j,k: two nodes in the network excluding i
- $f_{jk}(i)$: the proportion of shortest paths from vertex j to vertex k (> j) passing through node i

#### Implementation
1. Implement the function `calculate_betweenness_centrality`.This function calculates the betweenness centrality for all the nodes of the network and adds this measurement to each node.

**NB**: Use the functions `all_shortest_paths()` or `shortest_path()` from Networkx to compute the shortest simple paths in the graph.

2. The dataset has two connected components. Use the function `nx.connected_components` to get each one of them. Calculate the closeness centrality for each component, and explain your results. 

3. Indicate with the variable `field_betweenness_centrality` the field field where for each node you entered the value of betweeness centrality. It will allow the test function to check all of your centralities with a tolerance of 0.001


⚠️⚠️ Using `nx.betweenness_centrality` is forbidden, it is reserved for the tests

````python
from math import factorial

def calculate_betweenness_centrality(social_network: nx.Graph):
    """Calculate the betweenness centrality of each node in social_network and insert the value of each node in a new attribute"""
    n = len(social_network.nodes)
    all_shortest_paths = [ [0 for j in range(n)] for k in range(n)]
    for j in range(n) :
        for k in range(j, n) :
            if k in nx.shortest_path(social_network, source=j).keys() :
                all_shortest_paths[j][k] = list(nx.all_shortest_paths(social_network, source=j, target=k))

    den = factorial(n-1) / (2 * factorial(n-3))
    for node in social_network.nodes :
        sum = 0
        for j in social_network.nodes :
            for k in nx.shortest_path(social_network, source=j).keys() :
                if j != node and k != node and k > j :
                    paths_through_i = [path for path in all_shortest_paths[j][k] if node in path]
                    proportion = len(paths_through_i) / len(all_shortest_paths[j][k])
                    sum += proportion

        C_I = sum / den
        social_network.nodes[node]['betweenness centrality'] = C_I
        
        
        
    
field_betweenness_centrality = "betweenness centrality"

def test_hint_calculate_betweenness_centrality():
    """Tests the function calculate_betweenness_centrality"""
    if field_closeness_centrality == "":
        raise ValueError("Fill the variable field_closeness_centrality with the field where for each node we can find its centrality")
    social_network = load_unweighted_network(str(ROOT_DATA / "tp-marvel-dataset-subset.csv"))
    calculate_betweenness_centrality(social_network)
    centralities = [[social_network.nodes[i][field_betweenness_centrality],i] for i in range(social_network.number_of_nodes())]
    exp_centralities = [[v,k] for k,v in nx.betweenness_centrality(social_network).items()]
    centralities.sort(key=lambda x:-x[0])
    exp_centralities.sort(key=lambda x:-x[0])
    tolerance = 0.001
    for i,((c_pred, idx_pred), (c_true, idx_true)) in enumerate(zip(centralities, exp_centralities)):
        assert idx_pred==idx_true, f"Expecting the {i}-st highest betweenness centrality node to be {idx_true}"
        assert np.isclose(c_pred,c_true, rtol=tolerance, atol=tolerance), f"Expecting the {i}-st highest betweenness centrality to be {c_true}"

if NOTEBOOK:
    test_hint_calculate_betweenness_centrality()
````
